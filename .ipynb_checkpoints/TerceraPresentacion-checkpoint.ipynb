### Importación de librerías

```python
import serial 
import serial.tools.list_ports
import roboticstoolbox as rtb
import spatialmath as sm
import numpy as np
```

#### Identicación de puertos, configuración del serial.

```python
ports = serial.tools.list_ports.comports()

for port, desc, hwid in sorted(ports):
        print("{}: {} [{}]".format(port, desc, hwid))
```

```python
sPort = serial.Serial()
sPort.baudrate = 9600
sPort.port = '/dev/ttyACM0' # 'COM4'
try :
    sPort.open()
    sPort.is_open
finally:
    sPort.close()
```

#### Instancia del manipulador 

```python
thor = rtb.models.DH.ThorArm()
thor
```

```python
# Thor en configuración cero con todas las variables articulares en 0
thor.plot(thor.qz)
```

```python
# Thor en posición vertical
thor.plot(thor.qr)
```

#### Cinemática directa e inversa

```python
qx = np.deg2rad([45, 30, 50, 30, 10, 0])
qy = np.deg2rad([70, 35, 10, 0, 0, 0])
```

```python
Pose1 = thor.fkine(qx)
Pose1
```

```python
Pose2 = thor.fkine(qy)
Pose2
```

```python
thor.plot(qx)
```

```python
sol1 = thor.ikine_LM(Pose1)
np.rad2deg((sol1.q))
```

```python
thor.plot(sol1.q)
```

```python
thor.plot(qy)
```

```python
sol2 = thor.ikine_LM(Pose2)
np.rad2deg(sol2.q)
```

```python
t = np.arange(0, 2, 0.02);
traj = rtb.tools.mtraj(rtb.tools.trajectory.quintic, sol1.q, sol2.q, t);
```

```python
traj.q.shape
```

```python
traj.q
```

```python
thor.plot(traj.q)
```

```python
rtb.tools.trajectory.qplot(t, traj.q);
```

```python
T = thor.fkine(traj.q);
T
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(); ax = fig.add_subplot(111, projection="3d");
plt.plot(traj.q[:,0], traj.q[:,1], traj.q[:,2]);
```
